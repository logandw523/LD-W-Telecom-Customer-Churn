## Create a correlation heatmap using seaborn

```python
df_sub = df[['Gender', 'Age', 'Married', 'Number of Dependents','Number of Referrals','Tenure in Months', 'Total Charges', 'Total Refunds', 'Total Extra Data Charges', 'Total Long Distance Charges', 'Total Revenue',
'Avg Monthly Long Distance Charges', 'Avg Monthly GB Download', 'Online Backup', 'Premium Tech Support', 'Streaming TV','Payment Method', 'Monthly Charge','Customer Status', 'Churn Category']]
```

```python
df_sub.columns
```

```python
df_sub.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(20,10))

corr = df_sub.apply(lambda x: pd.factorize(x)[0]).corr()

corr_map = sns.heatmap(corr, annot = True, ax = ax,cmap='coolwarm',vmin=-1, vmax=1)
```

#### Looking at the colored heatmap, there are several conclusions to note about customer churn:

There is a weak but positive relationship between customer status and churn with age and married status. Maybe a competitor is targeting a certain subset of our customers by age.

There is a negative correlation betweeen customer status and churn with tenure in months. This can be expected as churn should decrease in chances with longer tenured customers.

Total charges has a negative correlation to churn. You would think that higher charges would be the main driver for customer churn. However monthly charges does have a slighly positive correlation with churn.

Payment method has a positive correlation with churn. Perhaps customers are dissatisfied with a payment method and needs to be investigated further.

Several premium offerings show relationships with churn such as online backup, premium tech support and streaming tv. Perhaps customers are not satisifed with these performances as causes of churn.

# 5. Create a categorical plot and interpret the results.

```python
sns.histplot(df['Monthly Charge'], bins = 25, kde = True)
```

```python
# Assign categories to Monthly Charge
df.loc[df['Monthly Charge'] < 20, 'Monthly Charge Category'] = 'Low Monthly Charge'
df.loc[(df['Monthly Charge'] >= 20) & (df['Monthly Charge'] <80), 'Monthly Charge Category'] = 'Moderate Monthly Charge'
df.loc[df['Monthly Charge'] >= 80, 'Monthly Charge Category'] = 'High Monthly Charge'
```

```python
df['Monthly Charge Category'].value_counts()
```

```python
# Create a categorical plot in seaborn using the price categories created above
f,ax = plt.subplots(figsize=(20,10))

sns.set(style="ticks")
catplot = sns.stripplot(data = df, x='Tenure in Months', y='Total Charges', hue='Monthly Charge Category', ax=ax)
```

The longer in tenure, the higher monthly charge a customer can expect. Most customers fall into moderate monthly charge between 20 and 80.
