# Data Wrangling

```python
# Convert Data Types
df_cust['Zip Code'] = df_cust['Zip Code'].astype(str)
df_cust['Latitude'] = df_cust['Latitude'].astype(str)
df_cust['Longitude'] = df_cust['Longitude'].astype(str)
df_cust['Gender'] = df_cust['Gender'].astype('category')
df_cust['Customer Status'] = df_cust['Customer Status'].astype('category')
```

```python
df_cust[['Zip Code','Latitude','Longitude','Gender','Customer Status']].info()
```

```python
# add State and Country Detail
df_cust['State'] = 'California'
df_cust['Country'] = 'United States'
```

# Data Cleaning

```python
# check unique values
for col in df_cust.columns:
    print('Column: {} / unique values: {}'.format(col, df_cust[col].unique()))
```

```python
# set max
pd.set_option('display.max_columns', 40)
```

```python
# check for nulls
df_cust.isnull().sum()
```

```python
# Fill objects and category NAs with 'Not Provided'
for col in ['Multiple Lines', 'Internet Type', 'Online Security','Online Backup','Device Protection Plan',
           'Premium Tech Support','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data','Churn Category','Churn Reason']:
    df_cust[col] = df_cust[col].fillna('Not Provided')
```

```python
df_cust.isnull().sum()
```

```python
df_cust[['Avg Monthly Long Distance Charges', 'Avg Monthly GB Download']].describe()
```

```python
# find median
df_cust[['Avg Monthly Long Distance Charges', 'Avg Monthly GB Download']].median()
```

```python
# fill float type NaNs with median
df_cust['Avg Monthly Long Distance Charges'] = df_cust['Avg Monthly Long Distance Charges'].fillna(
df_cust['Avg Monthly Long Distance Charges'].median())

df_cust['Avg Monthly GB Download'] = df_cust['Avg Monthly GB Download'].fillna(
df_cust['Avg Monthly GB Download'].median())
```

```python
# check for nulls
df_cust.isnull().sum()
```

```python
# Check for duplicates
df_cust.duplicated().sum()
```

```python
# Check for mixed types
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
        print(col)
```
